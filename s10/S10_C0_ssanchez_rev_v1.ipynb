$${\displaystyle \|{\vec {x}}\|_{p}={\sqrt[{p}]{|x_{1}|^{p}+|x_{2}|^{p}+...+|x_{n}|^{p}}}}$$

### Norma $L_2, p=2$

$${\displaystyle \|{\vec {x}}\|_{2}={\sqrt[{2}]{|x_{1}|^{2}+|x_{2}|^{2}+...+|x_{n}|^{2}}}}$$

### Norma $L_1,p=1$

$${\displaystyle \|{\vec {x}}\|_{1}={{|x_{1}|+|x_{2}|+...+|x_{n}|}}}$$

### Norma $L_\infty,p \to \infty $

$${\displaystyle \|{\vec {x}}\|_{\infty }=\max(|x_{1}|,|x_{2}|,...,|x_{n}|)}$$

###**Polinomio Característico**
R nos permite resolver el problema de autovalores y autovectores de una manera simple. Podemos crear una variable con dicha información asociada a una matriz; esta variable tendrá los auto-valores y los auto-vectores de la matriz.



---
**Problema de autovalores y autovectores.**

Si $T$ representa una transformación lineal de una espacio vectorial $V$ sobre un campo $F$ en si mismo y $\mathbf{v}$ es un vector diferente de cero en $V$, entonces  $\mathbf{v}$ es un vector propio de $T$ si $T(\mathbf{v})$ es un múltiplo escalar de $\mathbf{v}$. Matemáticamente, tendríamos que:

\begin{equation}
T(\mathbf{v})=\lambda \mathbf{v}
\end{equation}

Existe una correspondencia directa entre las matrices cuadradas $n\times n$ y las transformaciones lineales de un espacio vectorial n-dimensional en sí mismo, dada cualquier base del espacio vectorial. Por lo tanto, en un espacio vectorial de dimensión finita, es equivalente definir valores propios y vectores propios utilizando el lenguaje de matrices o el lenguaje de transformaciones lineales. Si $V$ es de dimensión finita, la ecuación anterior es equivalente a

\begin{equation}
A \mathbf{u}=\lambda \mathbf{u}
\end{equation}

donde $A$ es la matriz de representación de $T$ y $\mathbf{u}$ es el vector de coordenadas de $\mathbf{v}$

---

#**Listas**

Una lista es una secuencia o colección de diferentes ojetos de `R` como:

- Vectores
- Números.
- Caracteres
- Listas en si mismo.
- etc

Podemos considerar a una lista como un cntenedor de información correlacionada, bien estructurada y de facil lectura. Una lista acepta item de diferentes types; por el contrario, un vector (o matriz, que representa un vector multidimencional) no. Para crear una lista, simplemente llamamos al étodo `list()`, con el contenido dentro del paréntesis de la que queramos definir y separado por comas.

Vemos un ejemplo:

````
pelicula <- list("Toy Story", 1995, c("Animacion", "Aventura", "Comedia"))
````

Nótese en el codigo anterior que la variable `pelicula` contienen una lista de 3 objetos, los cuales corresponden a un `string`, un valor numérico y un vector de `strings`. Para ver el contenido de esta variable, simplemente la llamamos en una celda:

````
pelicula
````

Al igual que en `python`, una lista corresponde a una secuencia de elementos que representan indexados que comienzan desde el índice 1. Para acceder a uno de ellos utilizamos doble corchetes:

````
pelicula[[1]]
````

Notese que si consutamos la clase de objeto que nos devuel, corresponderá a un `string`:

````
class(pelicula[[1]])
````

##**Acceso a items de una lista**

Podemos obtener una parte de la lista usando el operador de corchetes simples `[]`. Este operador también se puede usar para obtener un único elemento en una posición especifica. Veamos el siguiente ejemplo:

````
pelicula[1]
````

Veamos que es una parte de la lista y que el elemente que devuelve es de la clase lista:

````
class(pelicula[1])
````

Así mismo, el número 2 devuelve el segundo elemento de la lista, si tal elemento existe:

````
pelicula[2]
````

````
class(pelicula[2])
````

Así mismo como realizabamos el seccionado de listas/vectores/arreglos-multidimensionales en python, en `R` podemos realizar esta operación mediante el operador `:`

````
pelicula[2:3]
````

En si mismo, este seccionado también corresponde a una lista

````
class(pelicula[2:3])
````

## **Listas nombradas**

Similar a un diccionario en `python`, en `R` podemos crear listas en donde a cada uno de los objetos que almacene se le asigne un nombre o palabra clave.

Veamos esto:

````
pelicula <- list(nombre = "Toy Story",
             año = 1995,
             genero = c("Animation", "Adventure", "Comedy"))
````

En este sentido, la lista `pelicula` posee objetos con nombres dentro de ella:

- El objeto `nombre` almacena un `string` con el nombre de la pelicula.
- El objeto `año` almacena un número con el año de la película.
- El objeto `genero` almacena un vector con los géneros dentro de los cuales se cataloga la película.

Para acceder a los valores de los objetos a través de los nombres, lo podemos realizar de tres formas:

- Mediante el operador $, de la forma `pelicula$nombre`, por ejemplo.
- Mediante el operador `[]`, de la forma `pelicula["nombre"]`.
- Mediante el operador de doble corchetes `[[]]`, de la forma `pelicula[["nombre"]]`.

Veamos lo anterior:

````
pelicula$año
````

````
pelicula$genero
````

````
pelicula$nombre
````

````
pelicula["nombre"]
````

````
pelicula["genero"]
````

````
pelicula["año"]
````

La diferencia entre ambos métodos de acceso es que el operador `$` devuelve el valor del objeto, mientra que `[]` una parte de la lista. 

Si accedemos mediante el operador de doble corchetes y el nombre del ítem, obtenemos directamente el objeto almacenado

````
pelicula[["nombre"]]
````

##**Adición, modificación y remoción de ítem de una lista**

Para adicionar un ítem nuevo a una lista, basta con asignar un valor a un campo nuevo que no exista. En el siguiente códico se adiciona un nuevo campo, `edad`, y se le asigna un valor numérico. En este caso se usa el operador doble corchetes ya que estamos haciendo referencia, directamente, a un miembro de la lista (en el cual agregamos un valor)

````
pelicula[["edad"]] <- 5
pelicula 
````

Nótese que se ha creado el nuevo campo de `edad`. Podemos reasignarle a este campo un valor de varias formas:

````
# Forma 1
pelicula['edad']<-6
pelicula
````

````
# Forma 2

pelicula[['edad']]<-8
pelicula
````

````
# Forma 3

pelicula$edad <- 10
pelicula
````

Si queremos remover un elmento, le reasinamos un valor `Null`:

````
pelicula[["edad"]] <- NULL
pelicula
````

##**Concatenación de listas**

La concateniación representa un proceso en el que unimos dos o más cosas en una secuencia. Para hacerlo, llamamos la función predefinica `c()`. 

Veamos un ejemplo mediante la definicio de dos listas:

````
pelicula_parte1 <- list(nombre = "Toy Story")
pelicula_parte2 <- list(año = 1995, genero = c("Animacion", "Aventura", "Comedia"))
````

Suppose we want a subset of the **`movies_Data`** data frame composed of movies from a given year forward (e.g. year 2000) if a selected variable is **recent**, or from that given year back if we select **old**. We can quite simply do that in R by doing this:

````
decada = 'reciente'

# If the decade given is recent...
if(decada == 'reciente' ){
    # Subset the dataset to include only movies after year 2000.
    subset(df, year >= 2000)
} else { # If not...
    # Subset the dataset to include only movies before 2000.
    subset(df, year < 2000)
}
````
